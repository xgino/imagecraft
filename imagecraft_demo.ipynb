```python
from PIL import Image
import matplotlib.pyplot as plt
import imagecraft as ic
```

```python
def show(title, original, processed):
    fig, ax = plt.subplots(1, 2, figsize=(8, 4))
    ax[0].imshow(original)
    ax[0].set_title("Original")
    ax[0].axis("off")

    ax[1].imshow(processed)
    ax[1].set_title(title)
    ax[1].axis("off")
    plt.show()
```

## JPG ImageTest

```python
# Load sample image
imgjpg = Image.open("./testimg/pied-cuckoo-9750790.jpg")

ic.get_image_info(imgjpg)
```

```python
# --- Conversions ---
png_img = ic.to_png(imgjpg, remove_bg=True, verbose=True)
show("PNG (bg removed)", imgjpg, png_img)

jpg_img = ic.to_jpg(imgjpg, verbose=True)
show("JPG (white bg)", imgjpg, jpg_img)
```

```python
# --- Enhancements ---
brighter = ic.adjust_brightness(imgjpg, 1.5, verbose=True)
show("Brightness x1.5", imgjpg, brighter)

contrast = ic.adjust_contrast(imgjpg, 1.5, verbose=True)
show("Contrast x1.5", imgjpg, contrast)

sharp = ic.adjust_sharpness(imgjpg, 2.0, verbose=True)
show("Sharpness x2.0", imgjpg, sharp)
```

```python
# --- Resize & crop ---
resized = ic.resize_with_aspect(imgjpg, (512, 818), verbose=True)
show("Resized (512x312)", imgjpg, resized)

ic.get_image_info(resized)
```

```python
# --- Presets ---
icon = ic.convert_icon(imgjpg, size=(128, 128), verbose=True)
show("Icon (128x128)", imgjpg, icon)

profile = ic.convert_profile_picture(imgjpg, size=(512, 512), verbose=True)
show("Profile Picture", imgjpg, profile)
```
